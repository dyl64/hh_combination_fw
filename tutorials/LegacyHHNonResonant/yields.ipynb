```python
import ROOT

def get_roorealvar_values(var):
    min_val = var.getMin()
    max_val = var.getMax()
    num_bins = var.getBins()
    bin_width = (max_val - min_val) / num_bins
    values = [min_val + (i + 0.5) * bin_width for i in range(num_bins)]
    return values
```

# bbll

```python
fi_bbll = ROOT.TFile("/afs/ihep.ac.cn/users/c/cyz/besfs/HH_comb/hh_combination_fw_new/run_final/output/workspace/nonres/bbll/0.root", "READ")
w_bbll = fi_bbll.Get("combWS")
```

```python
def get_yields_bbll(kl,k2v):
    iter = w_bbll.componentIterator()
    obj = iter.Next()
    w_bbll.var("mu").setVal(1)
    w_bbll.var("kl").setVal(kl)
    w_bbll.var("k2v").setVal(k2v)
    yields = 0
    while obj:
        if isinstance(obj, ROOT.RooRealSumPdf):
            observable = obj.getObservables(w_bbll.data("combData")).first()
            ind_yield = 0
            for i in get_roorealvar_values(observable):
                observable.setVal(i)
                nj = -1
                for j in obj.funcList():
                    nj += 1
                    fname  = j.GetName()
                    if 'L_x_ggF_' in fname or 'L_x_kl_' in fname or 'L_x_vbf' in fname:
                        ind_yield += j.getVal()
            # print(f"{obj.GetName()}: {ind_yield}")
            yields += ind_yield
        obj = iter.Next()
    # print(f"Total yields with kl={kl} k2v={k2v}: {yields}")
    return yields
```

# bbbb

```python
fi_bbbb = ROOT.TFile("/afs/ihep.ac.cn/users/c/cyz/besfs/HH_comb/hh_combination_fw_new/run_final/output/workspace/nonres/bbbb/0.root", "READ")
w_bbbb = fi_bbbb.Get("combWS")
```

```python
def get_yields_bbbb(kl,k2v):
    iter = w_bbbb.componentIterator()
    obj = iter.Next()
    w_bbbb.var("mu").setVal(1)
    w_bbbb.var("kl").setVal(kl)
    w_bbbb.var("k2v").setVal(k2v)
    yields = 0
    while obj:
        if isinstance(obj, ROOT.RooRealSumPdf):
            obj_name = obj.GetName()
            regime = obj_name.split("_")[-1]
            cleanname = obj_name.replace("_model","").replace("_channel_resolved","").replace("_channel_boosted","")
            if regime == "boosted":
                observable = w_bbbb.obj(f"obs_x_SR")
            else:
                observable = w_bbbb.obj(f"obs_x_{cleanname}")
            ind_yield = 0
            for i in get_roorealvar_values(observable):
                observable.setVal(i)
                if regime == "boosted":
                    ind_yield_i = obj.getVal() - w_bbbb.obj(f"DataPF_SR_scaleFactors_channel_boosted").getVal() * w_bbbb.obj(f"DataPF_SR_shapes_channel_boosted").getVal()
                else:
                    ind_yield_i = obj.getVal() - w_bbbb.obj(f"L_x_bkg_{cleanname}_overallSyst_x_StatUncert_channel_resolved").getVal() * w_bbbb.obj(f"binWidth_obs_x_{cleanname}_3_channel_resolved").getVal()
                ind_yield += ind_yield_i
            # print(f"{obj.GetName()}: {ind_yield}")
            yields += ind_yield
        obj = iter.Next()
    # print(f"Total yields with kl={kl} k2v={k2v}: {yields}")
    return yields
```
