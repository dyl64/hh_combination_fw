```python
def make_error_boxes(ax, xdata, ydata, xerror, yerror, facecolor='r',
                     edgecolor='none', alpha=0.5, hatch='none'):

    # Loop over data points; create box from errors at each point
    errorboxes = [Rectangle((x - xe[0], y - ye[0]), xe.sum(), ye.sum())
                  for x, y, xe, ye in zip(xdata, ydata, xerror.T, yerror.T)]

    # Create patch collection with specified colour/alpha
    pc = PatchCollection(errorboxes, facecolor=facecolor, alpha=alpha,
                         edgecolor=edgecolor, hatch=hatch)

    # Add collection to axes
    ax.add_collection(pc)

    # Plot errorbars
    #xerr = np.array(list(zip(xerror[0], xerror[1]))).T
    #yerr = np.array(list(zip(yerror[0], yerror[1]))).T
    #artists = ax.errorbar(xdata, ydata, xerr=xerr, yerr=yerr,
    #                      fmt='none', ecolor='k')

    #return artists
```

```python
for label_name, label_title in zip(['internal', 'preliminary', 'public'], ['Internal', 'Preliminary', '']) :
    
    fig, ax = plt.subplots(1, 1, figsize=(11, 14))

    my_channels = list(klambda_channels.keys())[::-1]

    # now set the y axis position for each channel
    # for each channel, we have both the observed and the expected
    # to take into account
    y_values = {}
    for ianalysis, analysis in enumerate(list(analysis_names.keys())) :
        vspace = 6.0*ianalysis
        y_values[analysis] = np.linspace(1+vspace, 12.0*(len(klambda_channels.keys()))+vspace, len(klambda_channels.keys()))
    titles  = [klambda_channels[channel]['title'] for channel in my_channels]

    # for each point, we need to set the channel name as tick label
    label_ticks = np.zeros(len(my_channels))
    for analysis, points in y_values.items() :
        label_ticks += points
    label_ticks = label_ticks / len(y_values.keys())


    # lines
    my_lines = 0.5*(label_ticks[1:]+label_ticks[:-1])
    my_width = 0.5*(label_ticks[1:]-label_ticks[:-1])
    my_lines = [my_lines[0]-my_width[0]*2.] + list(my_lines) + [my_lines[-1]+my_width[-1]*2.]
    # y lim
    ymin = y_values[list(y_values.keys())[0]][0] - (y_values[list(y_values.keys())[0]][1]-y_values[list(y_values.keys())[0]][0])/3.
    ymax = y_values[list(y_values.keys())[0]][-1] + (y_values[list(y_values.keys())[0]][1]-y_values[list(y_values.keys())[0]][0])*2.5
    
    ax.set_ylim(ymin, ymax)
    # set scale
    ax.set_xscale("linear")
    # x lim (hardcoded)
    ax.set_xlim(-12,32.5)
    # set ticks on the y axis
    ax.set_yticks(label_ticks)
    ax.set_yticklabels([u"%s" % klambda_channels[channel]['title'] for channel in my_channels], fontsize=21)
    
    # x axis
    majorticks = [-10, -5, 0, 5, 10, 15, 20]
    minorticks = [-12.5, -7.5, -2.5, 2.5, 7.5, 12.5, 17.5]
    ax.set_xticks(majorticks)
    ax.set_xticklabels(majorticks, fontsize=22)
    ax.set_xticks(minorticks, minor=True)
    for line in ax.xaxis.get_ticklines():
        line.set_markersize(10)
        line.set_markeredgewidth(1.5)
    for line in ax.xaxis.get_ticklines(minor=True):
        line.set_markersize(6)
        line.set_markeredgewidth(1.5)
    if iax == 0 :
        ax.set_xticklabels([], fontsize=22)
            
    # adjust frame width
    [i.set_linewidth(2.2) for i in ax.spines.values()]
        
    # y axis
    # remove tick lines
    for line in ax.yaxis.get_ticklines():
        line.set_markersize(0)
        line.set_markeredgewidth(0)
    for line in ax.yaxis.get_ticklines(minor=True):
        line.set_markersize(0)
        line.set_markeredgewidth(0)
            
    # do the lines
    for line in my_lines[1:] :
        ax.axhline(line, color='darkgrey', linewidth=1.5)
    
    # do the theory prediction
    for ichannel, channel in enumerate(my_channels) :
        ax.vlines(1, ymin=my_lines[ichannel], ymax=my_lines[ichannel+1], color='darkred', linewidth=2.5)
        
    for it, (t, tmarker) in enumerate(zip(["obs", "exp"], ['o', 's'])) :
        # pick central values (same for both 68% cl and 95% CL)
        central_values = [klambda_channels[channel]['68CL'][t][0] for channel in my_channels]
        # plot central values
        ax.errorbar(x=central_values, y=y_values[t], marker=tmarker, markersize=12, linestyle='none', color=analysis_names[t]['color'])
        # plot error bands
        for cl, ls, lw, lb in zip(['95CL', '68CL'], ['--', '-'], [3,5], ['95% CL', '68% CL']) :
            for ichannel, channel in enumerate(my_channels) :
                # get the intervals
                # there could be more than one
                my_intervals = klambda_channels[channel][cl][t][1]
                for iinterval, interval in enumerate(my_intervals) :
                    # get average between low and up edge
                    average = (interval[0]+interval[1])*0.5
                    # get low and upper distance from average
                    err_lo = np.abs(average-interval[0])
                    err_hi = np.abs(average-interval[1])
                    # plot the error bar
                    xerr = [[err_lo], [err_hi]]
                    eb = ax.errorbar(x=average, y=y_values[t][ichannel], capsize=4, xerr=xerr, yerr=0, ls=ls, lw=lw, color=analysis_names[t]['color'])
                
                    eb[0].set_linestyle('none')
                    eb[-1][0].set_linestyle(ls)
                    # change capsizes
                    for cap in eb[1] :
                        cap.set_markeredgewidth(4)
                        cap.set_markersize(12)
                if cl == '68CL' :
                    central = central_values[ichannel]
                    low_edges = np.array([interval[0] for interval in my_intervals])
                    err_los = np.abs(low_edges-np.ones(len(low_edges))*central)
                    idx_lo = np.argmin(err_los)
                    low_edge = low_edges[idx_lo]
                    up_edges = np.array([interval[1] for interval in my_intervals])
                    err_ups = np.abs(up_edges-np.ones(len(up_edges))*central)
                    idx_up = np.argmin(err_ups)
                    up_edge = up_edges[idx_up]
                    if channel == 'bbll' and t != 'obs' :
                        ax.text(18, y_values[t][ichannel]-1.25, "$%.0f^{%+.0f}_{%+.0f}$" % (central, np.abs(central-up_edge), (np.abs(central-low_edge)*-1)), fontsize=18)
                    else:
                        ax.text(18, y_values[t][ichannel]-1.25, "$%.1f^{%+.1f}_{%+.1f}$" % (central, np.abs(central-up_edge), (np.abs(central-low_edge)*-1)), fontsize=18)
                    if channel == 'ml' and t == 'obs' :
                        ax.text(22, y_values[t][ichannel]-1.25, r"$\cup$  $\kappa_{\lambda} \in [%+.1f, %.1f]$" % (my_intervals[0][0], my_intervals[0][1]), fontsize=18)
    
                
    # axis
    ax.set_xlabel("$\kappa_{\lambda}$", fontsize=24)

    # write legend
    legend_elements = [Line2D([0], [0], color=analysis_names[t]['color'], lw=5, ls='-', marker=tmarker, markersize=12, label=analysis_names[t]['title']+' 68% CL') for t, tmarker in zip(my_analysis_names, ['o', 's'])] \
                + [Line2D([0], [0], color=analysis_names[t]['color'], lw=3, ls='--', marker=None, label=analysis_names[t]['title']+' 95% CL') for t in my_analysis_names] \
                + [Line2D([0], [0], color='darkred', lw=2.5, label='SM prediction')]
    ax.legend(handles=legend_elements, loc='upper right', fontsize=20)
    # write atlas labels
    ax.text(0.04, 0.94, "ATLAS", style='italic', weight='bold', fontsize=23, transform=ax.transAxes)
    ax.text(0.17, 0.94, label_title, fontsize=23, transform=ax.transAxes)
    ax.text(0.04, 0.89, '$\sqrt{s}= 13$ TeV, 126—140 fb$^{-1}$', fontsize=23, transform=ax.transAxes)
    # write legend label
    ax.text(0.04, 0.84, r"$HH$ combination", transform=ax.transAxes, fontsize=22)
    plt.savefig("summary_HH_combination_and_channels_klambda_"+label_name+".pdf", bbox_inches='tight')
```
