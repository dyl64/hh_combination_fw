```python
def make_error_boxes(ax, xdata, ydata, xerror, yerror, facecolor='r',
                     edgecolor='none', alpha=0.5, hatch='none'):

    # Loop over data points; create box from errors at each point
    errorboxes = [Rectangle((x - xe[0], y - ye[0]), xe.sum(), ye.sum())
                  for x, y, xe, ye in zip(xdata, ydata, xerror.T, yerror.T)]

    # Create patch collection with specified colour/alpha
    pc = PatchCollection(errorboxes, facecolor=facecolor, alpha=alpha,
                         edgecolor=edgecolor, hatch=hatch)

    # Add collection to axes
    ax.add_collection(pc)

    # Plot errorbars
    #xerr = np.array(list(zip(xerror[0], xerror[1]))).T
    #yerr = np.array(list(zip(yerror[0], yerror[1]))).T
    #artists = ax.errorbar(xdata, ydata, xerr=xerr, yerr=yerr,
    #                      fmt='none', ecolor='k')

    #return artists
```

```python
fig, axs = plt.subplots(2, 1, figsize=(11, 13), gridspec_kw={'height_ratios': [3,2]})
plt.subplots_adjust(hspace=0)

axs = axs.flatten()

idx = 0
for iax, (ax, my_interpretation, my_interpretation_title) in enumerate(zip(axs, [combined_sm, combined_heft], ['$\kappa$ Interpretation', 'HEFT Intepretation'])) :
    my_dict[my_interpretation_title] = {}
    my_couplings = list(my_interpretation.keys())[::-1]
    
    npoints = {}
    for ianalysis, analysis in enumerate(list(analysis_names.keys())) :
        vspace = 6.0*ianalysis
        npoints[analysis] = np.linspace(1+vspace, 12.0*(len(my_interpretation.keys()))+vspace, len(my_interpretation.keys()))
    titles  = [my_interpretation[coupling]['title'] for coupling in my_couplings]

    label_ticks = np.zeros(len(my_interpretation.keys()))
    for analysis, points in npoints.items() :
        label_ticks += points
    label_ticks = label_ticks / len(npoints.keys())

    # lines
    my_lines = 0.5*(label_ticks[1:]+label_ticks[:-1])
    my_width = 0.5*(label_ticks[1:]-label_ticks[:-1])
    my_lines = [my_lines[0]-my_width[0]*2.] + list(my_lines) + [my_lines[-1]+my_width[-1]*2.]
    # y lim
    ymin = npoints[list(npoints.keys())[0]][0] - (npoints[list(npoints.keys())[0]][1]-npoints[list(npoints.keys())[0]][0])/3.
    if idx == 1 :
        ymax = npoints[list(npoints.keys())[0]][-1] + (npoints[list(npoints.keys())[0]][1]-npoints[list(npoints.keys())[0]][0])*1.5
    else :
        ymax = npoints[list(npoints.keys())[0]][-1] + (npoints[list(npoints.keys())[0]][1]-npoints[list(npoints.keys())[0]][0])*3.0
    
ax.set_ylim(ymin, ymax)
# set scale
ax.set_xscale("symlog")
# x lim (hardcoded)
ax.set_xlim(-10,50)
# set ticks on the y axis
ax.set_yticks(label_ticks)
ax.set_yticklabels([u"%s" % my_interpretation[coupling]['title'] for coupling in my_couplings], fontsize=21)
    
    # x axis
    majorticks = [-2, -1, 0, 1, 2, 4, 6,10, 20]
    minorticks = [-1.5, 0.5, 1.5, 3, 5, 8, 12,17]
    ax.set_xticks(majorticks)
    ax.set_xticklabels(majorticks, fontsize=22)
    ax.set_xticks(minorticks, minor=True)
    ax.set_xticklabels([], minor=True, fontsize=22)
    if iax == 0 :
        ax.set_xticklabels([], fontsize=22)
    # do the lines
    for line in my_lines[1:] :
        ax.axhline(line, color='darkgrey', linewidth=1)
    
    # do the theory prediction
    for icoupling, coupling in enumerate(my_couplings) :
        value = 1 if coupling == 'klambda' or coupling == 'k2V' or coupling == 'chhh' else 0
        ax.vlines(value, ymin=my_lines[icoupling], ymax=my_lines[icoupling+1], color='darkred')
        
    for it, t in enumerate(["obs", "exp"]) :
        # pick central values (same for both 68% cl and 95% CL)
        my_dict[my_interpretation_title][t] = {}
        central_values = [my_interpretation[coupling]['68CL'][t][0] for coupling in my_couplings]
        my_dict[my_interpretation_title][t]['x'] = [float(x) for x in central_values]
        for cl, ls, lb in zip(['95CL', '68CL'], ['--', '-'], ['95% CL', '68% CL']) :
                            
            err_lo = [np.abs(my_interpretation[coupling][cl][t][0]-my_interpretation[coupling][cl][t][1][0][0]) for coupling in my_couplings]
            err_hi = [np.abs(my_interpretation[coupling][cl][t][1][0][1]-my_interpretation[coupling][cl][t][0]) for coupling in my_couplings]
                
            xerr = np.array(list(zip(err_lo, err_hi))).T
                
            eb = ax.errorbar(x=central_values, y=npoints[t], capsize=4, xerr=xerr, yerr=0, ls=ls, marker='o', color=analysis_names[t]['color'], label=analysis_names[t]['title'] + ' ' + lb)
                
            my_dict[my_interpretation_title][t]['xerr_low_'+cl+"_"+str(1)] = [float(v) for v in err_lo]
            my_dict[my_interpretation_title][t]['xerr_high_'+cl+"_"+str(1)] = [float(v) for v in err_hi]
            eb[0].set_linestyle('none')
            eb[-1][0].set_linestyle(ls)
            # change capsizes
            for cap in eb[1] :
                cap.set_markeredgewidth(2)
                    
            my_dict[my_interpretation_title][t]['y'] = [float(v) for v in npoints[t]]
            my_dict[my_interpretation_title][t]['ylabel'] = [my_interpretation[coupling]['title'] for coupling in my_couplings]
                
            # for each coupling, write central value +/- 68% cl
            jdx = 0
            if cl == '68CL' :
                for v, lvl, lo, hi, coup in zip(central_values, list(label_ticks), err_lo, err_hi, my_couplings) :
                    if coup == 'ctthh' or coup == 'cgghh' :
                        ax.text(8, npoints[t][jdx]-1.25, "$%.2f^{%+.2f}_{%+.2f}$" % (v, hi, (lo*-1)), fontsize=17)
                    else :
                        ax.text(10, npoints[t][jdx]-1.25, "$%.1f^{%+.1f}_{%+.1f}$" % (v, hi, (lo*-1)), fontsize=17)

                    jdx += 1
# put black like above HEFT interpretation
axs[0].axhline(55, color='k', linewidth=1)
# write legend
legend_elements = [Line2D([0], [0], color=analysis_names[t]['color'], lw=1.5, ls='-', marker='o', label=analysis_names[t]['title']+' 68% CL') for t in my_analysis_names] \
                + [Line2D([0], [0], color=analysis_names[t]['color'], lw=1.5, ls='--', marker='o', label=analysis_names[t]['title']+' 95% CL') for t in my_analysis_names] \
                + [Line2D([0], [0], color='darkred', lw=2, label='SM prediction')]
#legend_elements = [legend_elements[0], legend_elements[2], legend_elements[1], legend_elements[3]] \
#                + [Line2D([0], [0], color='grey', lw=1.5, ls='--', label='SM prediction')]
axs[0].legend(handles=legend_elements, loc='upper right', fontsize=20)
# write nice text
axs[0].text(0.04, 0.53, '$\kappa$ Framework', transform=axs[0].transAxes, fontsize=21)
axs[1].text(0.04, 0.85, 'HEFT Framework', transform=axs[1].transAxes, fontsize=21)
# write atlas labels
axs[0].text(0.04, 0.92, "ATLAS", style='italic', weight='bold', fontsize=23, transform=axs[0].transAxes)
axs[0].text(0.17, 0.92, "Internal", fontsize=23, transform=axs[0].transAxes)
axs[0].text(0.04, 0.84, '$\sqrt{s}= 13$ TeV, 126140 fb$^{-1}$', fontsize=23, transform=axs[0].transAxes)
# write legend label
axs[0].text(0.04, 0.76, r"$HH$ $combination$", transform=axs[0].transAxes, fontsize=22)
plt.savefig("summary_HH_combination_sm_heft.pdf", bbox_inches='tight')
```
